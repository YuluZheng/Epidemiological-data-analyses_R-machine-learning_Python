```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split, GridSearchCV


# generate two class dataset
data = pd.read_csv("train-15.csv")
data.tail()
```

```python
data=data.drop(['No'],axis=1)
# Split target and features
X = data.drop(['IS'],axis=1)
X
y = data['IS']
y = y.astype(int)
y
print(X.shape)
print(y.shape)
```

```python
# train models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.linear_model import ElasticNet
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB

# logistic regression
model1 = LogisticRegression()
# knn
model2 = KNeighborsClassifier()
# RF
model3 = RandomForestClassifier()
# XGBoost
model4 = XGBClassifier()

# GaussianNB
model5=GaussianNB()

# neural network
model6 = MLPClassifier()

# fit model
model1.fit(X, y)
model2.fit(X, y)
model3.fit(X, y)
model4.fit(X, y)
model5.fit(X, y)
model6.fit(X, y)
```

```python
###########
## Tuning the hyper-parameters of an estimator

## Parameter estimation using grid search with cross-validation
###########
```

```python
from sklearn.model_selection import GridSearchCV

# model 1 - logistic regression

#tuned_params1 = {'C':[*np.linspace(0.01, 1.1,0.1)]}

C = np.arange(0.01, 1.1,0.1)
tuned_params1 = [{'C': C}]

GS1 = GridSearchCV(model1,tuned_params1,scoring="roc_auc",cv=10)
GS1.fit(X,y)
print(GS1.best_score_)
print(GS1.best_params_)

# best params:0.8976505720314609 {'C': 1.01}
```

```python
# model 2 - KNN
from sklearn.model_selection import GridSearchCV
k_range = range(1,15)
tuned_params2 = [{'n_neighbors': k_range}]

GS2 = GridSearchCV(model2, tuned_params2, scoring='roc_auc', cv=10)

GS2.fit(X,y)
print(GS2.best_score_)
print(GS2.best_params_)
```

```python
k_range_3rd = range(15, 35)
tuned_params2_3rd = [{'n_neighbors': k_range_3rd}]

GS2_3rd = GridSearchCV(model2, tuned_params2_3rd, scoring='roc_auc', cv=10)

GS2_3rd.fit(X, y)
print(GS2_3rd.best_score_)
print(GS2_3rd.best_params_)
#grid search best parameters for the neighbours are 310.8533523170131619
```

```python
# model 3 - RF
from sklearn.model_selection import cross_val_score
scorel = []
for i in range(0,200,10):
    rfc = RandomForestClassifier(n_estimators=i+1,random_state=25)          # n_estimator12131.....
                                 
    score = cross_val_score(rfc,X,y,cv=10).mean()
    scorel.append(score)
print(max(scorel),(scorel.index(max(scorel))*10)+1)   #101n_estimator1
# 1 - max(scorel)list-scorel2 - #list.index([object]) objectlist
# plt.figure(figsize=[20,5])    #205
# plt.plot(range(0,201,10),scorel)   #range 1-201,10scorel
# plt.show()
```

```python
tuned_params4 = [{'max_depth': np.arange(3, 10, 2), 'min_child_weight': np.arange(1, 6, 2)}]
GS4 = GridSearchCV(XGBClassifier(learning_rate=0.01, n_estimators=1000, silent=0, gamma=0.2, subsample=0.75, 
                                 objective='binary:logistic', reg_alpha=1, reg_lambda=1,seed=27),
                   param_grid=tuned_params4,
                   scoring='roc_auc', cv=10)
GS4.fit(X, y)
print(GS4.best_score_)
print(GS4.best_params_)

# 0.9235444878089849
# {'max_depth': 5, 'min_child_weight': 5}
```

```python
tuned_params4_1 = [{'max_depth': np.arange(5, 7, 2), 'min_child_weight': np.arange(6, 11, 2)}]
GS4_1 = GridSearchCV(XGBClassifier(learning_rate=0.01, n_estimators=1000, silent=0, gamma=0.2, subsample=0.75, 
                                 objective='binary:logistic', reg_alpha=1, reg_lambda=1,seed=27),
                   param_grid=tuned_params4_1,
                   scoring='roc_auc', cv=10)
GS4_1.fit(X, y)
print(GS4_1.best_score_)
print(GS4_1.best_params_)

# 0.9235650633315589
# {'max_depth': 5, 'min_child_weight': 6}
# Best parameter set
```
